```python
import os
import shutil
import random

def sample_files(input_folder, output_folder, sample_size=271):
    # Ensure the input folder exists
    if not os.path.exists(input_folder):
        print("The specified input folder does not exist.")
        return

    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get all files from the input folder
    files = [file for file in os.listdir(input_folder) if os.path.isfile(os.path.join(input_folder, file))]
    
    # Check if the folder contains enough files
    if len(files) < sample_size:
        print(f"Not enough files in the folder. The folder contains only {len(files)} files.")
        return

    # Select a random sample of files
    sampled_files = random.sample(files, sample_size)

    # Copy each sampled file to the output folder
    for file in sampled_files:
        src_path = os.path.join(input_folder, file)
        dst_path = os.path.join(output_folder, file)
        shutil.copy(src_path, dst_path)
        print(f"Copied {file} to {output_folder}")

# Specify the input and output folders
input_folder = 'images'
output_folder = 'sample_images_noise'

# Execute the function
sample_files(input_folder, output_folder)
```
